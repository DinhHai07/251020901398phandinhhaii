```python
import math

# Function to calculate the discriminant (delta)
def discriminant(a, b, c):
    return b**2 - 4*a*c

# Function to solve the quadratic equation
def solve_quadratic(a, b, c):
    delta = discriminant(a, b, c)

    if delta > 0:
        # Two real and distinct solutions
        x1 = (-b + math.sqrt(delta)) / (2 * a)
        x2 = (-b - math.sqrt(delta)) / (2 * a)
        print(f"The equation has two distinct real solutions: x1 = {x1}, x2 = {x2}")
    elif delta == 0:
        # One real solution
        x = -b / (2 * a)
        print(f"The equation has one real solution: x = {x}")
    else:
        # No real solutions, only complex
        realPart = -b / (2 * a)
        imaginaryPart = math.sqrt(abs(delta)) / (2 * a)
        print(f"The equation has two complex solutions: x1 = {realPart} + {imaginaryPart}i, x2 = {realPart} - {imaginaryPart}i")

# Input values from the user
a = float(input("Enter the value of a: "))
b = float(input("Enter the value of b: "))
c = float(input("Enter the value of c: "))

# Solve the quadratic equation
solve_quadratic(a, b, c)
```
